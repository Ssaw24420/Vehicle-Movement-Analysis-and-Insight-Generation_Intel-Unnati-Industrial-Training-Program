### importing library for  visualization 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### creating a dataframe

```python
data = {
    'timestamp': pd.date_range(start='2023-01-01', periods=100, freq='h'),  # Changed 'H' to 'h'
    'event': ['entry'] * 50 + ['exit'] * 50,
    'parking_lot': ['A'] * 25 + ['B'] * 25 + ['A'] * 25 + ['B'] * 25,
    'license_plate': [f'ABC{str(i).zfill(4)}' for i in range(100)]
}
df = pd.DataFrame(data)
```

###  extracting the date component from each timestamp

```python
df['date'] = df['timestamp'].dt.date
```

### defining a function to analyze and visualize vehicle movement patterns 

```python
def generate_movement_insights(df):
    # Calculate entry and exit counts per hour
    df['hour'] = df['timestamp'].dt.hour
    movement_counts = df.groupby(['hour', 'event']).size().unstack(fill_value=0)

    # Plot vehicle movement patterns
    plt.figure(figsize=(12, 6))
    movement_counts.plot(kind='bar', stacked=True)
    plt.title('Vehicle Movement Patterns')
    plt.xlabel('Hour of Day')
    plt.ylabel('Number of Vehicles')
    plt.xticks(rotation=0)
    plt.legend(title='Event')
    plt.show()

    return movement_counts
```

### analyzing  and visualizing unauthorized vehicle

```python
def generate_unauthorized_vehicle_insights(df, approved_vehicles):
    # Check for unauthorized vehicles
    unauthorized_vehicles = df[~df['license_plate'].isin(approved_vehicles['license_plate'])]

    # Plot unauthorized vehicle detection
    unauthorized_counts = unauthorized_vehicles.groupby('date').size()

    plt.figure(figsize=(12, 6))
    unauthorized_counts.plot(kind='bar')
    plt.title('Unauthorized Vehicle Detections')
    plt.xlabel('Date')
    plt.ylabel('Number of Unauthorized Vehicles')
    plt.xticks(rotation=45)
    plt.show()

    return unauthorized_vehicles
```

### generating insights from data

```python
approved_vehicles = pd.read_csv('C:\\Users\\Public\\The_Visionaries_PS-13\\stanford_dataset\\approved_vehicles.csv')  # CSV with columns: license_plate, owner_name, etc.

# Generate insights
movement_insights = generate_movement_insights(df)
unauthorized_vehicle_insights = generate_unauthorized_vehicle_insights(df, approved_vehicles)
```
